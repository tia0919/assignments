```python
import pandas as pd
```

```python
#create table showing the stats of football players based on 5 most important factors of choosing who you want on team
df = pd.read_csv('cleaned_players.csv')
for index, row in df.iterrows():
    
        
    df = df.reset_index()

    df = df[['first_name', 'second_name', 'element_type','goals_scored', 'total_points', 'assists', 'threat', 'selected_by_percent']]
```

```python
df
```

```python
#find the top 20 players based on how many goals are scored

goals_scored='goals_scored'
df_sorted=df.sort_values(by=goals_scored,ascending=False)
top_20=df_sorted.head(20)

print(top_20)
```

```python
#find the top 20 players based on how many total points are made

total_points='total_points'
df_sorted=df.sort_values(by=total_points,ascending=False)
top_20=df_sorted.head(20)

print(top_20)
```

```python
#find the top 20 players based on assists
assists='assists'
df_sorted=df.sort_values(by=assists, ascending=False)
top_20=df_sorted.head(20)

print(top_20)
```

```python
#find the top 20 players based on threat
threat='threat'
df_sorted=df.sort_values(by=threat, ascending=False)
top_20=df_sorted.head(20)

print(top_20)
```

```python
#find the top 20 players based on percentage selected
selected_by_percent='selected_by_percent'
df_sorted=df.sort_values(by=selected_by_percent, ascending=False)
top_20=df_sorted.head(20)

print(top_20)
```
